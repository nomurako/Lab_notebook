## ðŸ”§ Sensor Calibration Basics

Many sensors used in agriculture output **raw voltage**, **resistance**, or **digital values** that are not immediately meaningful. To use these in data analysis or control systems, we must convert them into **standard physical units** such as degrees Celsius (Â°C), kPa, or Âµmol/mÂ²/s.

### ðŸ“ Linear Calibration Formula

A typical calibration curve is defined by:

$$
y = a \cdot x + b
$$

Where:
- \( x \): raw sensor reading (e.g., voltage or digital value)
- \( y \): converted physical value (e.g., temperature)
- \( a \), \( b \): calibration coefficients (often provided by the manufacturer)

### ðŸ§ª Example: Temperature Sensor

A thermistor might output a voltage between 0 and 3.3V. Suppose its calibration is:

$$
\text{Temperature (Â°C)} = 25 \cdot \text{Voltage} - 10
$$

This means:
- A 0.8V reading â†’ \( 25 \times 0.8 - 10 = 10^\circ C \)
- A 1.0V reading â†’ \( 25 \times 1.0 - 10 = 15^\circ C \)

We will now implement this in code.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulated raw voltage data from a temperature sensor
voltage = np.array([0.8, 1.0, 1.2, 1.4, 1.6])

# Calibration equation: Temperature = 25 * Voltage - 10
temperature = 25 * voltage - 10

# Create a DataFrame for clarity
df_calib = pd.DataFrame({
    'Voltage (V)': voltage,
    'Temperature (Â°C)': temperature
})

print(df_calib)

# Plotting
plt.figure(figsize=(6, 4))
plt.plot(voltage, temperature, marker='o')
plt.title("Sensor Calibration: Voltage to Temperature")
plt.xlabel("Voltage (V)")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.show()
```

## ðŸ’§ Calculating Vapor Pressure Deficit (VPD)

### ðŸŒ¡ï¸ What is VPD?

**Vapor Pressure Deficit (VPD)** is the difference between the amount of moisture in the air and how much moisture the air can hold when it's saturated. 

It reflects the **evaporative demand** of the atmosphere and directly influences **transpiration** and **stomatal behavior** in plants.

---

### ðŸ”¬ Formula for VPD (in kPa)

1. Calculate **saturation vapor pressure** \( e_s \) from temperature \( T \) (Â°C):

$$
e_s = 0.6108 \cdot \exp\left( \frac{17.27 \cdot T}{T + 237.3} \right)
$$

2. Calculate **actual vapor pressure** \( e_a \) using relative humidity (RH):

$$
e_a = \frac{\text{RH}}{100} \cdot e_s
$$

3. Then, compute VPD:

$$
\text{VPD} = e_s - e_a = \left(1 - \frac{\text{RH}}{100}\right) \cdot e_s
$$

VPD is expressed in **kilopascals (kPa)**. A VPD of 0.4â€“1.2 kPa is generally favorable for many crops in greenhouses.

---


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulated hourly temperature and humidity data
temperature = np.array([20, 22, 25, 27, 30])  # Â°C
humidity = np.array([80, 75, 70, 60, 50])     # %

# Saturation vapor pressure (kPa)
e_s = 0.6108 * np.exp((17.27 * temperature) / (temperature + 237.3))

# Actual vapor pressure
e_a = (humidity / 100) * e_s

# VPD calculation
vpd = e_s - e_a

# Create DataFrame
df_vpd = pd.DataFrame({
    'Temperature (Â°C)': temperature,
    'Humidity (%)': humidity,
    'Saturation VP (kPa)': e_s,
    'Actual VP (kPa)': e_a,
    'VPD (kPa)': vpd
})

print(df_vpd)

# Plotting
plt.figure(figsize=(6, 4))
plt.plot(temperature, vpd, marker='o', color='darkred')
plt.title('VPD vs. Temperature (at decreasing RH)')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('VPD (kPa)')
plt.grid(True)
plt.show()
```

## â˜€ï¸ Estimating PAR from Solar Radiation

### ðŸ” What is PAR?

**Photosynthetically Active Radiation (PAR)** refers to the range of light wavelengths (400â€“700 nm) that plants use for **photosynthesis**. It is usually expressed in:

- **Âµmol photons / mÂ² / s**

---

### ðŸ“ Estimating PAR from Shortwave Radiation

If you have **shortwave radiation data** from sensors (measured in watts per square meter, W/mÂ²), you can estimate PAR using an **empirical conversion factor**.

### ðŸ’¡ Formula

$$
\text{PAR} = \text{Radiation} \times 2.0
$$

Where:
- Radiation is in **W/mÂ²**
- PAR is in **Âµmol/mÂ²/s**

> ðŸŒ¤ï¸ This factor (2.0 Âµmol/J) assumes full sunlight and standard atmospheric conditions.

---

### ðŸ“¦ Bonus: Daily Light Integral (DLI)

You can estimate **DLI** (total daily PAR) in mol/mÂ²/day by summing hourly PAR values:

$$
\text{DLI} = \frac{\sum \text{PAR (Âµmol/mÂ²/s)} \times 3600}{10^6}
$$

---


```python
# Simulated hourly radiation data (W/mÂ²)
radiation = np.array([200, 400, 600, 800, 1000])

# Convert to PAR (Âµmol/mÂ²/s) using factor of 2.0
par = radiation * 2.0

# Estimate DLI assuming 5 hours of these levels (simplified)
dli = (par * 3600).sum() / 1e6  # mol/mÂ²/day

# Create DataFrame
df_par = pd.DataFrame({
    'Radiation (W/mÂ²)': radiation,
    'PAR (Âµmol/mÂ²/s)': par
})

print(df_par)
print(f"\nEstimated Daily Light Integral (DLI): {dli:.2f} mol/mÂ²/day")

# Plot
plt.figure(figsize=(6, 4))
plt.plot(radiation, par, marker='o', color='green')
plt.title('PAR vs. Radiation')
plt.xlabel('Radiation (W/mÂ²)')
plt.ylabel('PAR (Âµmol/mÂ²/s)')
plt.grid(True)
plt.show()
```
